```python
from dreadnode.agent import Agent, MultiAgent
```

```python
researcher = Agent(
    name="Researcher",
    description="Specializes in finding and summarizing information from the web.",
    tools=[...],
)

coder = Agent(
    name="Coder",
    description="Specializes in writing and debugging code.",
    tools=[...], # Code execution tools
)

# Define the Director and assign workers
research_team_director = MultiAgent(
    name="ResearchTeamDirector",
    description="Manages a team of researchers and coders to produce technical reports.",
    workers=[researcher, coder],
    instructions="""
    Your goal is to produce a comprehensive technical report.
    1.  Use the Researcher to gather information.
    2.  Use the Coder to write any necessary code examples.
    3.  Synthesize the information into a final report.
    """,
)

# Run the multi-agent system
final_report = await research_team_director.run("Produce a report on the performance of different sorting algorithms.")
```
