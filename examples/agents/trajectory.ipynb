```python
import dreadnode
from dreadnode import Agent, tool
from rich import print

dreadnode.configure(server="https://platform.dreadnode.io", workspace="trajectories", organization="dreadnode")

@tool
def generate_joke(joke: str) -> str:
    """Write a joke yourself, then use the generate_joke tool to save it."""
    return f"Here's a joke for you: {joke}"


agent = Agent(
    name=f"ExampleAgent",
    description="An example agent for demonstration purposes.",
    model="groq/openai/gpt-oss-120b",
    tools=[generate_joke]
)

result = await agent.run("Please generate 5 different jokes for me.")
```

```python
print(result.trajectory_to_nemo())
```

```python
print(result.steps)
```
